```python
import insilicoICH
from insilicoICH import load_phantom

IPH_ME = load_phantom(11.5)
IPH_ME.insert_lesion('IPH', seed=42, mass_effect=True, intensity=60, volume=5)
EDH_ME = load_phantom(11.5)
EDH_ME.insert_lesion('EDH', seed=42, mass_effect=True, intensity=60, volume=10)
SDH_ME = load_phantom(11.5)
SDH_ME.insert_lesion('SDH', seed=42, mass_effect=True, intensity=60, volume=10)

IPH_noME = load_phantom(11.5)
IPH_noME.insert_lesion('IPH', seed=42, mass_effect=False, intensity=60, volume=5)
EDH_noME = load_phantom(11.5)
EDH_noME.insert_lesion('EDH', seed=42, mass_effect=False, intensity=60, volume=10)
SDH_noME = load_phantom(11.5)
SDH_noME.insert_lesion('SDH', seed=42, mass_effect=False, intensity=60, volume=10)
```
